```python
# load libraries
import tonic
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load SHD
SHD = tonic.datasets.SHD(save_to = "data",
                         train = True)
```

```python
x_data = []
y_data = []

for i in SHD:
    x_data.append(i[0])
    y_data.append(i[1])
    
x_data = np.array(x_data, dtype = 'object')
y_data = np.array(y_data)
```

```python
# get a single sample
print(f"y data      : {SHD[0][1]}")
print(f"x data      : {SHD[0][0]}")
print(f"first index :{SHD[0]}")
print(f"speaker     :{SHD.speaker}")
```

```python
speaker = 0
digit = 2

print(f"digit: {y_data[np.where(SHD.speaker == speaker)][np.where(y_data[np.where(SHD.speaker == speaker)] == digit)]}")
print(f"speaker: {SHD.speaker[np.where(SHD.speaker == speaker)]}")
```

```python
# average digit of each speaker
# speakers (11) with a sub array of digits (20)

average_speaker_digit = [[] for _ in range(max(SHD.speaker) + 1)]

for speaker in average_speaker_digit:
    for digit in np.unique(y_data):
        speaker.append([digit])
    
```

```python
# create single digit x from speaker y

speaker = 1
digit = 2
index = 0

plt.scatter(x_data[np.where(SHD.speaker == speaker)][np.where(y_data[np.where(SHD.speaker == speaker)] == digit)][index]["t"],
            x_data[np.where(SHD.speaker == speaker)][np.where(y_data[np.where(SHD.speaker == speaker)] == digit)][index]["x"],
            s = 0.5)

plt.ylim(0, 700)
plt.ylabel("input neurons")
plt.xlabel("time")
plt.title("SHD example")
plt.show()
```

```python
# display combined digit x for speaker y

speaker = 1
digit = 5

for index in range(y_data[np.where(SHD.speaker == speaker)][np.where(y_data[np.where(SHD.speaker == speaker)] == digit)].shape[0]): 
    # iterating through all digit examples for specified speaker
    plt.scatter(x_data[np.where(SHD.speaker == speaker)][np.where(y_data[np.where(SHD.speaker == speaker)] == digit)][index]["t"],
                x_data[np.where(SHD.speaker == speaker)][np.where(y_data[np.where(SHD.speaker == speaker)] == digit)][index]["x"],
                s = 0.1)

plt.ylim(0, 700)
plt.ylabel("input neurons")
plt.xlabel("time")
plt.title(f"SHD combined for digit {digit} spoken by speaker {speaker}")
plt.show()
```

```python
# combined digit x for speaker y

speaker = 1
digit = 5

x, t = [], []
combined_images = []

for index in range(y_data[np.where(SHD.speaker == speaker)][np.where(y_data[np.where(SHD.speaker == speaker)] == digit)].shape[0]): 
    # iterating through all digit examples for specified speaker
    t += list(x_data[np.where(SHD.speaker == speaker)][np.where(y_data[np.where(SHD.speaker == speaker)] == digit)][index]["t"])
    x += list(x_data[np.where(SHD.speaker == speaker)][np.where(y_data[np.where(SHD.speaker == speaker)] == digit)][index]["x"])

# idx works on np.array and not lists.
list1 = np.array(t)
list2 = np.array(x)
idx   = np.argsort(list1)

list1 = np.array(list1)[idx]
list2 = np.array(list2)[idx]

plt.scatter(list1, list2,
            s = 0.1)

plt.ylim(0, 700)
plt.ylabel("input neurons")
plt.xlabel("time")
plt.title(f"SHD combined for digit {digit} spoken by speaker {speaker}")
plt.show()
```
