```python
import tonic
import matplotlib.pyplot as plt

# Load the SHD training dataset
shd_train = tonic.datasets.SHD(save_to='./data', train=True)

# Print dataset size
print(f"Number of training samples: {len(shd_train)}")

# Load one sample (events, target label)
events, label = shd_train[0]

# Extract event timestamps and neuron IDs
spike_times = events['t']  # Spike timestamps
neuron_ids = events['x']  # Neuron IDs

# Plot spike raster for one sample
plt.figure(figsize=(10, 5))
plt.scatter(spike_times, neuron_ids, s=2)
plt.xlabel("Time (ms)")
plt.ylabel("Neuron ID")
plt.title(f"SHD Spike Raster Plot for Digit: {label}")
plt.show()
```
